```python
import pandas as pd
import math
from PIL import Image
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from skimage.feature import hog
from skimage.color import rgb2grey
from skimage.filters import prewitt_h,prewitt_v, prewitt, sobel
import sys
import cv2
from matplotlib.colors import hsv_to_rgb
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
```

1. LOAD IMAGE

```python
def get_image(my_img,my_ext='.png'):
    img = cv2.imread("./"+str(my_img)+str(my_ext),cv2.IMREAD_UNCHANGED)
    #we convert back the image from BGR to RGB
    img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
my_image = get_image("pokemon1",".jfif")
plt.imshow(my_image)
```

REMOVE BACKGROUND

```python
def get_contrasted(image, type="dark", level=3, my_phi=1,my_theta=1):
    maxIntensity = 255.0 # depends on dtype of image data
    phi = my_phi
    theta = my_theta

    if type == "light":
        newImage0 = (maxIntensity/phi)*(image/(maxIntensity/theta))**0.5
        newImage0 = newImage0.astype(np.uint8)
        return newImage0
    elif type == "dark":
        newImage1 = (maxIntensity/phi)*(image/(maxIntensity/theta))**level
        newImage1 = newImage1.astype(np.uint8)
        return newImage1
```

```python
def remove_contours_2(my_image_name):
    
    gray_img = cv2.cvtColor(my_image_name, cv2.COLOR_BGR2GRAY)
    gray_img=get_contrasted(gray_img)
    
    _,thresh = cv2.threshold(gray_img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    thresh = cv2.bitwise_not(thresh)
    element = cv2.getStructuringElement(shape=cv2.MORPH_RECT, ksize=(10, 10))
    morph_img = thresh.copy()
    cv2.morphologyEx(src=thresh, op=cv2.MORPH_CLOSE, kernel=element, dst=morph_img)
    contours,_ = cv2.findContours(morph_img,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    areas = [cv2.contourArea(c) for c in contours]
    sorted_areas = np.sort(areas)
    number_of_areas=sorted_areas.shape[0]
        
    cnt=contours[areas.index(sorted_areas[-1])] #the biggest contour
    area = cv2.contourArea(cnt)
    perimeter = cv2.arcLength(cnt,True)
    epsilon = 0.001*cv2.arcLength(cnt,True)
    approx = cv2.approxPolyDP(cnt,epsilon,True)
    
    mask = np.zeros(gray_img.shape,np.uint8)
    cv2.drawContours(mask,[cnt],0,255,-1)
    final_image = cv2.bitwise_and(my_image_name,my_image_name, mask=mask)
    
    return final_image
```

```python
my_image_wobkgd=remove_contours_2(my_image)
plt.imshow(my_image_wobkgd)
```

RESIZE to large pixels

```python
def resize_image(img, my_scale):
    print('Original Dimensions : ',img.shape)
    scale_percent = my_scale # percent of original size
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)
    dim = (width, height)
    # resize image
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_NEAREST)
    print('Resized Dimensions : ',resized.shape)
    
    return resized
```

```python
my_pxl_image=resize_image(my_image_wobkgd,10)
plt.imshow(my_pxl_image)
```

REDUCE number of colours

```python
def reduce_colours (my_image,number_of_bins):
    print("the number of final colours is {0}".format(number_of_bins**3))
    img_r,img_g,img_b = cv2.split(my_image)
    my_labels=np.arange(0,256,255/(number_of_bins-1)).astype(np.uint8)
    new_colour_map_red=np.reshape(pd.cut(img_r.flatten(),number_of_bins, labels=my_labels, include_lowest=True),(img_r.shape[0],img_r.shape[1]))
    new_colour_map_green=np.reshape(pd.cut(img_g.flatten(),number_of_bins, labels=my_labels, include_lowest=True),(img_g.shape[0],img_g.shape[1]))
    new_colour_map_blue=np.reshape(pd.cut(img_b.flatten(),number_of_bins, labels=my_labels, include_lowest=True),(img_b.shape[0],img_b.shape[1]))
    final_image=cv2.merge([new_colour_map_red,new_colour_map_green,new_colour_map_blue])
    return final_image
```

```python
my_pxl_color_image=reduce_colours(my_pxl_image,8)
plt.imshow(my_pxl_color_image)
```

BIND IT TOGETHER

```python
def pixel_generator(my_image, ext='.jpg',pxl_pc=10,color_bin=8):
    my_image = get_image(my_image,ext)
    my_image_wobkgd=remove_contours_2(my_image)
    my_pxl_image=resize_image(my_image_wobkgd,pxl_pc)
    my_pxl_color_image=reduce_colours(my_pxl_image,color_bin)
    fig,(ax1,ax2,ax3,ax4)=plt.subplots(1,4,figsize=(20,20))
    ax1.imshow(my_image)
    ax2.imshow(my_image_wobkgd)
    ax3.imshow(my_pxl_image)
    ax4.imshow(my_pxl_color_image)
    ax1.set_title("original image")
    ax2.set_title("removed contour")
    ax3.set_title("pixel image with all colours")
    ax4.set_title("final image with minimum colours")
```

```python
pixel_generator("pikachu",".jfif",10,6)
```
